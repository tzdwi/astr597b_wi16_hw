```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#We'll be using LSD, so let's make sure to only run this on Magneto, and only run it once we've run the following:

#LSD environment variables
import os
os.environ['NWORKERS'] = '8'
os.environ['LSD_CACHEDIR'] = os.environ['HOME']+'/tempdir'
os.environ['LSD_TEMPDIR'] = os.environ['HOME']+'/tempdir'
os.environ['LSD_DB'] = os.environ['HOME']+'/lsddb:/astro/apps6/opt/lsddb'

#Let's import LSD
from lsd import DB
from lsd.bounds import beam, rectangle
```

```python
db = DB(os.environ["LSD_DB"])
```

```python
field = db.query('ra, dec, g, r FROM sdss WHERE (g-r<0.4) & (r<=22.0) & (r>=20.0) & (type==6)').fetch()
```

```python
#properties in each r bin to make this easier
blue_dict = {'r_min':20.,'r_max':20.66,'n_min':98,'n_max':267.}
green_dict = {'r_min':20.66,'r_max':21.33,'n_min':107.,'n_max':304.}
red_dict = {'r_min':21.33,'r_max':22.0,'n_min':102.,'n_max':330.}

#Map a color to a dictionary
dict_dict = {'blue':blue_dict,'green':green_dict,'red':red_dict}
```

```python
def norm_hist(color):
    color_dict = dict_dict[color]
    color_field = field[(field['r']>color_dict['r_min'])&(field['r']<=color_dict['r_max'])]
    
    #Coverage is 360 deg in RA and 110 deg in Dec, so we'll want 720 by 220 bins to get 0.5 deg resolution
    color_hist,color_ra,color_dec = np.histogram2d(color_field['ra'],color_field['dec'],bins=[720,220])
    color_hist = 4.*color_hist #to get from stars per 0.25 square deg. to stars per square deg.
    #Now we need to select the range of densities used in the plot, subtract the minimum off, then normalize so that imshow knows what its doing
    color_hist[(color_hist<color_dict['n_min'])|(color_hist>color_dict['n_max'])]=0. #get rid of values outside of our range
    color_hist[color_hist!=0.]-=color_dict['n_min'] #where we actually have a value, subtract off the minimum
    
    return color_hist/color_hist.max() #normalize
```

```python
from matplotlib.colors import LogNorm
im = np.flipud(np.fliplr(np.array([norm_hist('red'),norm_hist('green'),norm_hist('blue')]).transpose()))
#gets the image in the correct shape

fig, ax = plt.subplots(figsize=(16,8))
img = ax.imshow(im,extent=(max(field['ra']),min(field['ra']),min(field['dec']),max(field['dec'])),norm=LogNorm())
 
ax.set_xlim(230.,110.)
ax.set_ylim(-5.,60.)
ax.set_xlabel('RA')
ax.set_ylabel('Dec')
```
